## Notification aux utilisateurs au sein d'un groupe


Ce script permet d'envoyer une notification à tous les utilisateurs d'un groupe d'un portail ArcGIS

#### Connexion au SIG :

```python
from arcgis.gis import GIS
gis = GIS("home")
```

#### Lister les utilisateurs du groupe

```python
#Trouver le groupe 
groupe = gis.groups.search('title:Notifications', max_groups=15) #changer le titre du groupe
groupe
```

```python
#Lister les membres du groupe
membres = groupe[0].get_members()

#formattage des données
liste_membres = []
for clef, membre in membres.items() :
    if clef !='owner':
        liste_membres.append(membre[0])
liste_membres
```

#### Envoi de la notification aux membres du groupe

```python
gis.users.send_notification(liste_membres, "Je suis vraiment trop forte",#titre du message
                            "Le script marche comme un grand et t'envoie ce message pour te dire que ton client va être content", #contenu du message
                            "email" #à retirer pour avoir une notification sur le portail
                           )
#doc : https://developers.arcgis.com/python/api-reference/arcgis.gis.toc.html#arcgis.gis.UserManager.send_notification
```
