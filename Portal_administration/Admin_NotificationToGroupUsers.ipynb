## Notification aux utilisateurs au sein d'un groupe


Ce script permet d'envoyer une notification Ã  tous les utilisateurs d'un groupe d'un portail ArcGIS

#### Connexion au SIG :

```python
from arcgis.gis import GIS
gis = GIS("home")
```

#### Lister les utilisateurs du groupe

```python
#Find group
group = gis.groups.search('title:Notifications', max_groups=15) #changer le titre du groupe
group
```

```python
#List group members
members = group[0].get_members()

#Formating data
members_list = []
for role, member in members.items() :
    if role !='owner':
        members_list.append(member[0])
members_list
```

#### Envoi de la notification aux membres du groupe

```python
gis.users.send_notification(members_list, "Notification title",
                            "Notification content", 
                            "email" #delete this line for builtin notification
                           )
#doc : https://developers.arcgis.com/python/api-reference/arcgis.gis.toc.html#arcgis.gis.UserManager.send_notification
```
